```python
df.drop(['oldbalanceOrg','newbalanceDest'],axis=1,inplace=True)
```

## **Feature Engineering**

```python
df=df[df.nameDest.str.startswith('M')==False]
```

```python
df=df[df.nameOrig.str.startswith('M')==False]
```

```python
from category_encoders import TargetEncoder
tenc=TargetEncoder()
df['nameOrig']=tenc.fit_transform(df['nameOrig'],df['isFraud'])
```

```python
from category_encoders import TargetEncoder
tenc=TargetEncoder()
df['nameDest']=tenc.fit_transform(df['nameDest'],df['isFraud'])
```

```python
temp=pd.get_dummies(df['type'])
```

```python
df=df.merge(temp,left_index=True,right_index=True)
```

```python
df.drop('type',axis=1,inplace=True)
```

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=9,contamination=0.1,novelty=True)
clf.fit(df)
```

```python
x=clf.predict(df)
df['is_outlier']=x
df=df[df['is_outlier']==1]
df.drop('is_outlier',axis=1,inplace=True)
```

```python
X=df.drop('isFraud',axis=1)
y=df['isFraud']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2)
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
X_train,y_train=sm.fit_resample(X_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

## **Model Creation**

```python
#Creating baseline model
from sklearn.ensemble import GradientBoostingClassifier
m1=GradientBoostingClassifier()
m1.fit(X_train,y_train)
pred=m1.predict(X_test)
```

```python
from sklearn.metrics import recall_score,f1_score,accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score
```

```python
recall_score(y_test,pred)
```

```python
f1_score(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

```python
precision_score(y_test,pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot()
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
pred=xgb.predict(X_test)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot()
```

```python
recall_score(y_test,pred)
```

```python
from lightgbm import LGBMClassifier
lgbm=LGBMClassifier()
lgbm.fit(X_train,y_train)
pred=lgbm.predict(X_test)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot()
```

```python
recall_score(y_test,pred)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```
